<h3>Removing Outliers</h3>

Using Z-score Technique

```python
from scipy.stats import zscore
z = np.abs(zscore(df))
z.shape
```

```python
z = pd.DataFrame(z, columns=df.columns)
z
```

Any value above 3 in z-score table is considered an outlier

```python
print(np.where(z>3))
```

```python
df_new = df[(z<3).all(axis=1)]
print("Old Shape", df.shape)
print("New Shape", df_new.shape)
print("Total Dropped Rows", df.shape[0] - df_new.shape[0])
```

```python
df_new
```

```python
from sklearn.preprocessing import power_transform
x = df_new.iloc[:,:-1]
y = df_new.iloc[:,-1]
```

<h4>Trnasformation to remove skewness</h4>

```python
x = power_transform(x, method='yeo-johnson')
x
```

```python
x_df = pd.DataFrame(x, columns=df_new.columns[:-1])
x_df.skew()
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(x)
x
```

```python
x_df = pd.DataFrame(x, columns=df_new.columns[:-1])
x_df.mean()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
from sklearn.metrics import accuracy_score
```

<h3>Making predictions</h3>

```python
for i in range(0,1000):
    x_train, x_test, y_train, y_test = train_test_split(x,y,random_state=i, test_size=0.20)
    lr.fit(x_train, y_train)
    pred_train = lr.predict(x_train)
    pred_test = lr.predict(x_test)
    if round(accuracy_score(y_train, pred_train)*100, 1) == round(accuracy_score(y_test, pred_test)*100,1):
        print("At random state", i, "The model performs very well")
        print("At random state", i , ":")
        print("Training Accuracy is :" ,round(accuracy_score(y_train, pred_train)*100, 1))
        print("Testing Accuracy is :" ,round(accuracy_score(y_test, pred_test)*100, 1))
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=986)
```

```python
lr.fit(x_train, y_train)
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_pred, y_test))
```

```python
lss = accuracy_score(y_test, y_pred)
lss
```

Cross Validation

```python
from sklearn.model_selection import cross_val_score
for j in range(2,10):
    lssscore = cross_val_score(lr,x,y, cv=j)
    lsc =lssscore.mean()
    print("At CV :", j)
    print("Cross Validation score is : -", lsc*100)
    print("Accuracy sicre is : -", lss*100)
```

```python
from sklearn.metrics import roc_curve, auc
fpr, tpr, thresholds = roc_curve(y_pred, y_test)
roc_auc = auc(fpr,tpr)
plt.figure()
plt.plot(fpr, tpr, color='darkorange', lw=10, label='ROC Curve (area=0.2%f)' %roc_auc)
plt.plot([0,1], [0,1], color='navy',lw=10, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive rate')
plt.title('Reciever Operating Characteristics')
plt.legend(loc='lower right')
plt.show()
```
